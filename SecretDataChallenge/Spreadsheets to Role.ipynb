```python
import graphlab
```

```python
data = graphlab.SFrame.read_csv("spreadsheets/SurveyAnalysis/data/SurveyData_Numeric_Cleaned1.csv/")
```

```python
data2 = data["Q3.3_1", "Q3.3_2", "Q3.3_3", "Q3.3_4", "Q3.3_5", "Q3.3_6", "Q12.9"]
```

```python
data2
```

```python
data2['functionality'] = data2['Q3.3_1'] + data2['Q3.3_2'] + data2['Q3.3_3'] + data2['Q3.3_4'] + data2['Q3.3_5'] + data2['Q3.3_6']
```

```python
data2['count'] = data2['functionality'].apply(lambda x: x.count("1"))
```

```python
data2
```

```python
data2.remove_columns(['functionality'])
```

```python
def area(x):
    if x == "1":
        x = "Sales"
    if x == "2":
        x = "Marketing"
    if x == "3":
        x = "Operations"
    if x == "4":
        x = "Engineering"
    if x == "5":
        x = "Research"
    if x == "6":
        x = "Finance"
    if x == "7":
        x = "Distribution"
    if x == "8" or x == "0":
        x = "Other"
    return x
```

```python
data2['role'] = data2['Q12.9'].apply(area)
```

```python
data2
```

```python
data2.sort('count')
```

```python
data2.show(view="Bar Chart", x="role", y="count")
```

```python
import graphlab.aggregate as agg
role_count = data2.groupby(key_columns='role', operations={'count': agg.COUNT()})
```

```python
role_count
```

```python
role_count.sort('count')
```

```python
role_count.show(view="Bar Chart", x="role", y="count")
```
